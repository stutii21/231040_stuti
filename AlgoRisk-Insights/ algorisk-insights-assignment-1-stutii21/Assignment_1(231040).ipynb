In general, most investors will prefer to have a lower correlation as this mitigates risk in their portfolios of different assets or securities being impacted by similar market conditions. However, risk-seeking investors or investors wanting to put their money into a very specific type of sector or company may be willing to have higher correlation within their portfolio in exchange for greater potential returns.

```python
import yfinance as yf
import matplotlib.pyplot as plt

stocks = ['MARUTI.NS','INFY.NS']
data = yf.download(stocks, start='2023-04-01', end='2024-03-31')['Adj Close']
rets = data.pct_change()
corr_day = rets.rolling(window=20).corr()
corr_matrix = corr_day.iloc[-len(stocks):]

plt.figure(figsize=(10, 10))
plt.imshow(corr_matrix, cmap='RdYlGn', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation='vertical')
plt.yticks(range(len(corr_matrix)), corr_matrix.columns)
plt.suptitle('Rolling Stock Correlations Heat Map', fontsize=15, fontweight='bold')
plt.show()
corr_matrix
```

```python
data3 = pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/RELIANCE.NS.csv")
data3
```

```python
import matplotlib.pyplot as plt
stock_data=pd.DataFrame(data3)
start_date = '2023-04-01'
end_date = '2024-03-31'
filtered_data = stock_data[(stock_data['Date'] >= start_date) & (stock_data['Date'] <= end_date)]

# 10-day SMA
sma_10 = []
for i in range(len(filtered_data)):
    if i < 9:
        sma_10.append(None)
    else:
        sma_10.append(filtered_data['Close'].iloc[i-9:i+1].mean())
filtered_data['10-daysma'] = sma_10
# 20-day SMA
filtered_data['20-daysma'] = filtered_data['Close'].rolling(window=20).mean()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(filtered_data['Date'], filtered_data['Close'], label='Closing Price')
plt.plot(filtered_data['Date'], filtered_data['10-daysma'], label='10-daysma')
plt.plot(filtered_data['Date'], filtered_data['20-daysma'], label='20-daysma')
plt.title('Stock Price with 10-day and 20-day Simple Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
